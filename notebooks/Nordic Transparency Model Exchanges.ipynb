```python
# Show exchanges on map.
import folium
from folium import plugins
m = folium.Map(location=(62,17), zoom_start=3.5)

# Exclude areas
d = ['Norway, Norway MBA, Stattnet CA', 'Sweden, Sweden MBA, SvK CA']

# Add areas to map.
for area in ex_areas:
    
    
    #if from_area not in d and to_area not in d:
    folium.GeoJson(ee.areas[ee.areas['Meaning']==area], name=area).add_to(m)

for idx, row in n_ex[1:].iterrows():
    
    # Inverts coordinates
    #coords = list(n.lines.at[idx,'geometry'].coords)
    #if coords[0][0] < 40:
        #coords = [(x[1], x[0]) for x in coords]
    #name = row['name']
    #l_m = row['l_m']
    # From center
    from_area = row['in_domain.mrid']
    #print(from_area)
    from_point = ee.areas[ee.areas['Meaning'] == from_area]['coords'].values[0]
    #print(from_point)
    
    #from_point = ee.areas[ee.areas['Meaning'] == from_area]['geometry'].representative_point()
    #print(from_point)
    
    
    # To center
    to_area = row['out_domain.mrid']
    to_point = ee.areas[ee.areas['Meaning'] == to_area]['coords'].values[0]
    #print(to_point)
    
    coords = [from_point, to_point]
    coords = [(x[1], x[0]) for x in coords]
    
    if from_area not in d and to_area not in d:
        folium.PolyLine(coords, color='purple').add_to(m)
    
# Add folim control
folium.LayerControl().add_to(m)
    
m
```
